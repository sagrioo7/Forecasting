```python
import pandas as pd
import numpy as np

data=pd.read_excel("CocaCola_Sales_Rawdata.xlsx")
data
```

```python
quarter=['Q1','Q2','Q3','Q4']
n=data['Quarter'][0]
n[0:2]
```

```python
data['quarter']=0
for i in range(42):
    n=data['Quarter'][i]
    data['quarter'][i]=n[0:2]
```

```python
dummy=pd.DataFrame(pd.get_dummies(data['quarter']))
dummy
```

```python
df=pd.concat((data,dummy),axis=1)
df
```

```python
df['t']=np.arange(0,42)
df['t_square']=df['t']*df['t']
df.head(42)
```

```python
log_sales=np.log(df['Sales'])
df['log_Sales']=log_sales
df
```

```python
train=df.head(37)
test=df.tail(4)
train
```

```python
test
```

```python
df.Sales.plot()
```

```python
import statsmodels.formula.api as smf 

linear = smf.ols('Sales~t',data=train).fit()
pred_linear =  pd.Series(linear.predict(pd.DataFrame(test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_linear))**2))
rmse_linear
```

```python
Quad = smf.ols('Sales~t+t_square',data=train).fit()
pred_Quad = pd.Series(Quad.predict(test[["t","t_square"]]))
rmse_Quad = np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_Quad))**2))
rmse_Quad
```

```python
Exp = smf.ols('log_Sales~t',data=train).fit()
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(test['t'])))
rmse_Exp = np.sqrt(np.mean((np.array(test['Sales'])-np.array(np.exp(pred_Exp)))**2))
rmse_Exp
```

```python
add_sea = smf.ols('Sales~Q1+Q2+Q3+Q4',data=train).fit()
pred_add_sea = pd.Series(add_sea.predict(test[['Q1','Q2','Q3','Q4']]))
rmse_add_sea = np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

```python
add_sea_Quad = smf.ols('Sales~t+t_square+Q1+Q2+Q3+Q4',data=train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(test[['Q1','Q2','Q3','Q4','t','t_square']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad 
```

```python
Mul_sea = smf.ols('log_Sales~Q1+Q2+Q3+Q4',data = train).fit()
pred_Mult_sea = pd.Series(Mul_sea.predict(test))
rmse_Mult_sea = np.sqrt(np.mean((np.array(test['Sales'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea
```

```python
Mul_Add_sea = smf.ols('log_Sales~t+Q1+Q2+Q3+Q4',data = train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(test))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(test['Sales'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea 
```

```python
data = {"MODEL":pd.Series(["rmse_linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea","rmse_Mult_add_sea"]),"RMSE_Values":pd.Series([rmse_linear,rmse_Exp,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_sea,rmse_Mult_add_sea])}
data
```

```python
table_rmse=pd.DataFrame(data)
table_rmse
```

```python
"""WE WILL USE Multiplicative Additive Seasonality AS ITS RMSE VALUE IS THE LEAST 
SO ITS THE BEST MODEL TO FORECAST"""
```

```python
import pandas as pd
import numpy as np

data1=pd.read_excel("Airlines+Data.xlsx")
data1
```

```python
data1["Date"] = pd.to_datetime(data1.Month,format="%b-%y")
#look for c standard format codes

# Extracting Day, weekday name, month name, year from the Date column using 
# Date functions from pandas 

data1["month"] = data1.Date.dt.strftime("%b") # month extraction
data1["year"] = data1.Date.dt.strftime("%Y") # year extraction

#Walmart["Day"] = Walmart.Date.dt.strftime("%d") # Day extraction
#Walmart["wkday"] = Walmart.Date.dt.strftime("%A") # weekday extraction
```

```python
month =['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'] 
month
```

```python
month_dummies = pd.DataFrame(pd.get_dummies(data1['month']))
month_dummies
```

```python
df = pd.concat([data1,month_dummies],axis = 1)
df
```

```python
df["t"] = np.arange(1,97)
df["t_squared"] = df["t"]*df["t"]
df.columns
df["log_Passengers"] = np.log(df["Passengers"])
df
```

```python
Train = df.head(84)
Test = df.tail(12)
Train
```

```python
Test
```

```python
df.Passengers.plot()
```
